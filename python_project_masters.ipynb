```python
df = pd.read_csv(r'C:\Users\Admin\Downloads\SIMPLILEARN\projects\project_1.csv', parse_dates=['Created Date', 'Closed Date'])
```

```python
df['Request_Closing_Time'] = df['Closed Date'] - df['Created Date']
```

```python
df['Status'].value_counts().plot(kind='bar',figsize=(6,6))
plt.show()
```

```python
from scipy.stats import ttest_1samp
mean = np.mean(df)
print(mean)
tset, pval = ttest_1samp(df)
print(pval)
if pval < 0.05:    
   print(" we are rejecting null hypothesis")
else:
  print("we are accepting null hypothesis")
```
